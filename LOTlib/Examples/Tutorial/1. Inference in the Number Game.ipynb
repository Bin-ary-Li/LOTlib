LOTlib is a Python 2.x library that provides a generic framework for representing and reasoning over "language of thought" (LOT) models. LOTlib can be used to model compositional learning of concepts over a wide range of domains. This tutorial is intended to answer the questions: what are some basic concepts we should be familiar with, and how do we implement grammars in LOTlib to model an arbitrary domain?

### Grammars, Primitives, & Hypotheses
LOTlib is fundamentally based around the concept of abstract *grammars*, or sets of rules. When an individual attempts to learn patterns from input data, we can model this individual as having an internal grammar of primitives. In this introduction, input data will be lists of integers, and learners' primitives will include integers themselves as well as arithmatic operations such as addition, subtraction, and exponentiation. 

In this example, we will consider rules as mapping one set of integers to another. With a set of domain entities taken as input, a learner should be able to learn the expression that was used to generate these items. Our domain for this example will be the integers $n \in \mathbb{N}$ such that $1 \geq n \geq 100$. The expression  $n \rightarrow 2n$  would map to the set of even numbers in our domain:  $\{2,4,6,...,100\}$ 

One application of a *grammar* is to generate arbitrary expressions, including both *syntactic* and *primitive* expressions. In an English-language-syntax grammar, syntactic expressions would include labels such as 'sentence', 'noun phrase', or 'adjective'; primitive expressions would be the actual words in the lexicon, such as 'bottom', or 'sandy'.

Using the primitives $1, 2, 3, +, -, *, ^$, a grammar is capable of generating compose infinitely long rules (or *"sentences"*). This is what we mean by *"compositional learning of concepts"*. E.g. we could have the number $1000$ by appling the rule $Expr \longrightarrow Expr + Expr$ 1000 times, with the second right-hand-side product always reducing to '1'. In other words, "one plus one, repeated a thousand times, equals 1000", or: $1000 = (1 + (1 + (1 + (1 + (1 + (. . .))))))$

Let us consider the following grammar, which lets us generate mathematical expressions:


$
\textbf{First Rule}        \\
Start \longrightarrow Expr \\\ \\ 
\textbf{Terminals}         \\
Expr \overset{2.0}{\longrightarrow} n \\
Expr \overset{1.0}{\longrightarrow} 1 \\
Expr \overset{1.0}{\longrightarrow} 2 \\
... \\
Expr \overset{1.0}{\longrightarrow} 10 \\\ \\ 
\textbf{Syntax} \\
Expr \overset{1.0}{\longrightarrow} Expr + Expr \\
Expr \overset{1.0}{\longrightarrow} Expr - Expr \\
Expr \overset{1.0}{\longrightarrow} Expr * Expr \\
Expr \overset{1.0}{\longrightarrow} Expr^{Expr} \\
$

It's worth noting here that our grammar is *probabilistic*, which means that each rule has an associated probability weight. Because $Expr \overset{4.0}{\longrightarrow} n$ has a weight of 4 and the rule $Expr \overset{2.0}{\longrightarrow} 2$ has weight 2, given an 'Expr' node the first rule is twice as likely to be called as the second rule.  Any sentence generated by this grammar can be part of a *hypothesis*, which associates posterior, prior, and likelihood probabilities with an expression.

Using this grammar, we can generate expressions such as: 

$
2n \sim \{2,4,6,8,...\}\ \ \ \ [even\ numbers] \\
8\ \sim \{8\}\ \ \ \ [eight] \\
2^{n} \sim \{2,4,8,16,32,64\}\ \ \ \ [powers\ of\ two] \\ 
2^{n}-1\ \sim \{1,3,7,15,31,63\}\\
$

We can get these expressions by beginning in our grammar with a 'Start' node and using rules with 'Expr' on the left-hand side to generate more nodes until all our nodes are leaves.


## Example
It is not the case that a hypothesis is equivalent to its corresponding $C$. Rules can be used to spontaneously *generate* hypotheses, but a grammar alone does not enumerate all its derivable hypotheses. The grammar above has an infinite *space* of possible hypotheses as there are rules such as $Expr \longrightarrow Expr * Expr$, which could recurse infinitely as in $(Expr * Expr * Expr * Expr *\ .\ .\ .)$. 
Children are capable of learning abstract rules using sparse, noisy, exclusively positive data. For example, a child might be repeatedly told that ${1,3,5,7,9,11,15,...}$ are odd numbers, but never explicitly learn that 13 is an odd number. There might even be an error or two, where perhaps the child mistook a set of cardinality 3 to have 2 items, and was told the set size was odd.

In this case, the child's input data could map to the set $D = \{1,1,1,2,3,3,3,...,7,9,9,9,11,11,11,15,15,...\}$, while their hypothesis maps to $C = \{1,3,5,7,9,11,...,99\}$. The correct hypothesis for this data would correspond to the most likely rule used to generate this data. The following rule was used to generate $D$:
$$ n \longrightarrow 2n - 1 $$

The value we care to calculate is the posterior probability $p(h|D)$ of each hypothesis, "the chance that this hypothesis was what generated the data we're seeing". We use two other values in calculating this - the prior probability of the hypothesis $p(h)$, or "the chance that our hypothesis would be generated from our grammar" - and the likelihood of the data being generated by the hypothesis $p(D|h)$, or "the chance that given this hypothesis, we would generate this data".

**Likelihood** is important since it lets us prefer the best hypothesis between multiple possible choices. E.g. input data $D = \{4,8,32\}$ could be generated by either hypothesis, $n \longrightarrow 2^{n}$ or $n \longrightarrow 2n$. The first hypothesis ('powers of two') maps to the set $\mathcal{H} = \{2,4,8,16,32,64\}$; each datum has likelihood $1/6$ of being generated, since the length of $\mathcal{H}$ is 6. The second hypothesis ('even numbers') maps to $\mathcal{H} = \{2,4,6,8,10,...,98,100\}$, so each datum has likelihood $1/50$ of being generated.

As we can see from this, it is more likely that $D = \{4,8,32\}$ was generated by the first hypothesis than the second. 

**Prior** probability is important as well, since this will let us prefer simpler hypotheses.. If we consider the two equivalent hypotheses, e.g. $2n = times(2,n)$ & $(2n+1-1) = minus(plus(times(2,n),1,1)$, we may prefer the first hypothesis if it is more likely to be generated than the second. 


The following code implements a model of this scenario to stochastically find the best hypotheses.

Setting up a grammar
--------------------

We would also like to consider sets of integers generated using set-theoretic operations. For example, we might have the input $D = \{2,4,5,8,10,15,16,20,25,30,32,...\}$. In plain English, this is "numbers that are either a power of two or a multiple of five"; in formal terms: $D = \{n \mid \exists x . n = 2^{x}\} \cup \{n \mid \exists x . n = 5x \}$. Let's expand our grammar to include set operations:

$
\textbf{First Rule}       \\
Start \longrightarrow Set          \\\ \\ 
\textbf{Primitives}       \\
Expr \overset{4.0}{\longrightarrow} n    \\
Expr \overset{2.0}{\longrightarrow} 1    \\
Expr \overset{2.0}{\longrightarrow} 2    \\
Expr \overset{2.0}{\longrightarrow} 3    \\\ \\ 
\textbf{Expressions}       \\
Expr \longrightarrow Expr + Expr    \\
Expr \longrightarrow Expr - Expr    \\
Expr \longrightarrow Expr * Expr    \\
Expr \longrightarrow Expr^{Expr}    \\
Func \longrightarrow \lambda n . Expr(n) \\\ \\
\textbf{Sets} \\
Set \longrightarrow Set \cup Set \\
Set \longrightarrow Set \cap Set \\
Set \longrightarrow Set \setminus Set \\
Set \longrightarrow Range(Num, Num) \\
Set \longrightarrow Map(Func, Set) \\
$


We must also define a new class for the new way we are thinking about numbers. In this case, any expressions such as $2^{n} - 1$ or $3n$ will be applied to some other set, for example the set produced by the rule $Range(1,100)$. When sets are evaluated, they will inevitably bottom out with evaluations of $Range(Num, Num)$, with various set rules applied such as $\cup$ & $\cap$. 

Using this new grammar, we can generate expressions such as: 

$
2n \cup 5n \sim \{2,4,5,6,8,...\}\ \ \ \ [even\ or\ power\ of\ five] \\
8 \cup 17 \sim \{8,17\}\ \ \ \ [eight\ or\ seventeen] \\
2^{n} \setminus {16} \sim \{2,4,8,32,64,...\}\ \ \ \ [powers\ of\ two\ except\ 16] \\ 
Range(1,100)\ \ \ \ [range\ from\ 1\ to\ 100]
$

```python
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~ Import Stuff ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
%matplotlib inline
import sys
from IPython.display import clear_output
from pylab import *
from LOTlib import lot_iter
from LOTlib.Inference.MetropolisHastings import MHSampler
from LOTlib.Miscellaneous import logsumexp, exp
from LOTlib.Examples.NumberGame.Model import *

#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
#~~~~~~~~~~~~~~~~~~~~~~~~ Setting up our LOT hypothesis grammar ~~~~~~~~~~~~~~~~~~~~~~~~#
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
del grammar
grammar = Grammar()

# Sets
grammar.add_rule('START', '', ['SET'], 1.)
# grammar.add_rule('SET', 'union_', ['SET', 'SET'], 1.)
# grammar.add_rule('SET', 'setdifference_', ['SET', 'SET'], 1.)
# grammar.add_rule('SET', 'intersection_', ['SET', 'SET'], 1.)

# Range of numbers, e.g. [1,100] (numbers 1 through 100)
# grammar.add_rule('SET', 'range_set_', ['EXPR', 'EXPR', 'bound=100'], 10.)
# grammar.add_rule('SET', 'range_set_', ['1', '100', 'bound=100'], 8.)
grammar.add_rule('rng', 'range_set_', ['1', '100', 'bound=100'], 8.)

# Mapping expressions over sets of numbers
grammar.add_rule('SET', 'mapset_', ['FUNC', 'rng'], 3.)
# grammar.add_rule('SET', 'mapset_', ['FUNC', 'SET'], 3.)
grammar.add_rule('FUNC', 'lambda', ['EXPR'], 1., bv_type='EXPR', bv_p=1.0)

# Expressions
# grammar.add_rule('EXPR', 'plus_', ['EXPR', 'EXPR'], 1.)
# grammar.add_rule('EXPR', 'minus_', ['EXPR', 'EXPR'], 1.)
grammar.add_rule('EXPR', 'times_', ['EXPR', 'EXPR'], 1.)
grammar.add_rule('EXPR', 'ipowf_', ['EXPR', 'EXPR'], 1.)

# Terminals
for i in range(0, 10):
    grammar.add_rule('EXPR', str(i), None, 1.)

print 'Now we have a grammar!'
```

# So now we have the grammar, let's do some inference . . . (TODO: improve this bit)

On input $D = \{2, 4, 16, 64\}$, a learner should be able to recognize that these are all powers of 2. In other words, the hypothesis which was used to generate these numbers is:
$$ n \longrightarrow 2^{n} $$
This rule states that for some subset of the integers in the range $[1,100]$, there are mappings to other integers $f(n)=2^{n}$ in the same range. The full set of numbers correctly generated by this rule is: $C = \{2,4,8,16,32,64\}$.

However, we can imagine other hypotheses which could generate this same data. For example, perhaps the rule is for powers of 2 except for the numbers 8 & 32:
$$ A = \{n \mid n \in 2^{n}\} $$
$$ C = A - \{8,32\} $$
$D$ could have also been generated by an 'even numbers only' rule:
$$ n \longrightarrow 2n $$
or even by a rule with no constraints on the domain, where $C_{h} = [1,100]$:
$$ n \longrightarrow n $$

The first hypothesis has two advantages for which we should prefer it over the other three. 

###### Hypothesis complexity
Equation 1 is syntactically simpler than equation 2 (TODO: equation labels), which means it is more likely to be found by a learner searching through spaces of possible hypotheses. 

###### Likelihood (of generating data)
Second,  is more likely to generate $D$ than equations 3 or 4. E.g. for equation 4, $C_{h} = {1,2,...100}$, which means the probability of generating $n$

```python
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~ Initial parameters ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
# Input data
# data = [1,1,1,3,3,3,3,3,5,5,7,7,9,9,11,11,11,15,15,17,17,17,17,19,19,19]  # for 2n - 1
##data = [3, 9, 17]       # for 2^n + 1
data = [2, 2, 2, 2, 2, 8, 8, 8, 8, 8, 8, 16, 16, 16, 32, 32, 32, 32, 32, 32, 64, 64, 64]
# Range of number in domain => [1, domain]
domain = 100
# Number of hypotheses to generate
num_iters = 5000
# Noise parameter
alpha = .99

#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~ Randomly generate samples ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
# Initialize an empty set to store our hypotheses
hypotheses = set()

# Iterate for the specified number of times
for _ in lot_iter(xrange(num_iters)):
    if _ % 2000 == 0:
        clear_output()
        print '\nGenerating %i hypotheses...\n' % _                # Printing statement #
        sys.stdout.flush()
    
    # Generate a hypothesis & add it to our set
    t = NumberGameHypothesis(grammar, domain=domain, alpha=alpha)
    t.compute_posterior(data)                        # Get prior, likelihood, posterior #
    hypotheses.add(t)
    
# Estimate normalizing constant Z 
Z = logsumexp([h.posterior_score for h in hypotheses])


#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
#~~~~~~~~~~~~~~~~~~~~~~ Printing stuff for best 10 hypotheses ~~~~~~~~~~~~~~~~~~~~~~~~~~#
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#

best_hypotheses = sorted(hypotheses, key=lambda x: x.posterior_score)
best_hypotheses.reverse()

clear_output()
print '\nGenerated %i hypotheses in total' % num_iters
print str(len(hypotheses)) + ' unique hypotheses\n'
print '================================================================'

for i in range(10):
    h = best_hypotheses[i]
    # h_in_domain is the set of items this hypothesis maps to; e.g. 2^n => {2,4,8,16,...}
    h_in_domain = [item for item in h() if item <= domain]
    
    # Create figure
    figure(num=None, figsize=(8, 1.5), dpi=80, facecolor='w', edgecolor='k')
    hist(h_in_domain, bins=domain, range=(1,domain))
    
    title('Hypothesis: ' + str(h) +
          '\nPosterior: %.5f' % exp(h.posterior_score - Z) +
          '\nPrior: %.3f' % h.prior +
          '\nLikehd: %.3f' % h.likelihood)

    
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
#~~~~~~~~~~~~~~~~~~~ Remember, the ideal hypothesis is:  2n + 1 ~~~~~~~~~~~~~~~~~~~~~~~~#
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
```

The learner estimates a *hypothesis* to explain the mapping from input to output: $n \longrightarrow f(n)$. Since it also must be the case that $f(n) \in [1,100]$, $C$ is always a subset of $[1,100]$. A *hypothesis* is actually a tree of rules generated using our grammar. For any domain, we need to define a hypothesis class with a method for computing the likelihood. Since likelihood is computed as "the probability that the data was generated by our hypothesis", or $p(D|h)$, this method will tell the hypothesis how to evaluate itself. In our domain, all hypotheses are lambda expressions that take a single input 'n', mapped to all numbers in the domain $[1,100]$ - for example, $2^{n} - 1$ maps to ${1,3,7,15,31,63}$.
